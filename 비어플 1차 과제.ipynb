1. BMW Cars Dataset
이 데이터를 고른 이유 : 
이 데이터셋은 전 세계가 사랑하는 자동차 브랜드이면서 저의 드림카인 BMW를 다루고 있어 실제 자동차 시장 분석, 가격 예측 모델링, 마케팅 분석 등 다양한 현실 문제로 확장 가능함.
데이터 크기가 충분히 커서 EDA(탐색적 자료 분석)이 가능한 규모임.
또한 대부분이 분석하기 쉬운 표 형식의 데이터이어서 데이터 전처리가 좋을 것 같다고 생각함.
데이터의 특징 :
연식, 가격, 엔진사양, 변속기 등 다양한 변수들이 존재함. 이러한 변수들을 통해 차량의 기본 정보를 알 수 있기도 하고 가격 정보, 성능 및 사양 특징 등을 알 수 있음. 예를 들어 연식별 가격 변화 추세 분석이 가능하고 연료 타입/ 변속기 등으로 가격 차이 분석이 가능함. 이 데이터셋은 범주형과 숫자형이 적절히 섞인 형태이며 전처리가 쉬운 구조로 되어 있어 분석의 난이도를 조절하기가 쉬운 장점이 있음.

2. 공연장/문예회관 등 문화시설 추이
이 데이터를 고른 이유 :
아무래도 문화시설 수는 국민의 여가 활동, 문화 접근성, 문화 인프라의 발전 수준을 보여주는 중요한 사회지표이고 문화인으로서 문화시설 수는 문화적 기회가 많다는 것이기 때문에 관심이 많아서 이 데이터를 고름.
데이터의 특징 :
이 지표는 다수의 연도 데이터를 포함하고 있어서 시간 흐름에 따른 데이터 변화 분석이 가능함. 이에 시계열 분석에 매우 적합한 형태임. 데이터를 보면 공연장, 문예회관 두 시설 모두 전반적으로 지속 증가하는 추세를 보이는 것을 알 수 있음. 이에 이 지표는 문화 인프라 현황을 객관적으로 이해하게 해주는 매우 실용적인 통계로 말할 수 있음.
